# Project Problem Statement: Optimizing Brand Campaign ROI Across the Creator Economy

## Context
A multinational consumer brand allocates millions of dollars annually to influencer marketing across multiple platforms (Instagram, YouTube, TikTok). The marketing team needs to move away from vanity metrics (like follower counts) and deeply understand actual consumer conversion behavior across different audience demographics, creator niches, and content formats.

## The Business Pain Points
The brand wants to uncover which precise factors drive consumer decisions, repeat interactions, and direct product sales. Specifically, they want to analyze:
* **Platform Efficiency**: How consumer engagement shifts between channels.
* **Niche Performance**: Which creator domains (e.g., Tech, Fashion, Fitness) yield the highest transaction volume relative to cost.
* **Campaign Mechanics**: The actual performance impact of tactical variations like exclusive promo codes or giveaways.
* **Seasonality**: How purchasing velocity spikes during critical commercial periods.

## Overarching Analytical Question
> **"How can consumer brand managers leverage creator campaign performance metrics to identify high-converting audience segments, optimize multi-platform ad spend, and systematically maximize Return on Ad Spend (ROAS)?"**

```python
import numpy as np

#synthetic fault injection to make sure data contains nan values
np.random.seed(42)


mask_sales = np.random.rand(len(df)) < 0.05


df.loc[mask_sales, 'product_sales'] = np.nan


print("--- Verified Data Quality Check (Permanent Nulls Injected) ---")
df.isnull().sum()
```

```python
df['product_sales'] = df.groupby('influencer_category')['product_sales'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#snake casing( everything in lowercase with n underscore in b/w)
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
#creating a cloumn influencer tiers
labels= [ 'Micro-influencer', 'Mid-tier', 'Macro-influencer', 'Meg-influencer']
df['influencer_tier']= pd.qcut(df['estimated_reach'], q=4, labels= labels)
df[['estimated_reach', 'influencer_tier']].head(5)
```

```python
# map
platform_overhead_mapping = {
    'YouTube': 14,    # Long-form video production takes about 14 days
    'TikTok': 7,      # Short-form video takes about 7 days
    'Instagram': 5,   # Carousel/Image posts take 5 days
    'Twitter': 2      # Text/Thread campaigns take 2 days
}

df['estimated_production_days'] = df['platform'].map(platform_overhead_mapping)


df[['platform', 'estimated_production_days']].head(10)
```

```python
# 1. Intentionally create a completely redundant text column for good practice
df['has_engagement_tracking'] = 'Yes'


print(" Column List with Redundant Field Added ")
print(df.columns.tolist())
```
